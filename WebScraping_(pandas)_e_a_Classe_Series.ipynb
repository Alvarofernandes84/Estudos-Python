```python
# Nessa situação estou importando a classe "Series" que existe na biblioteca "pandas".
# Ela é específica para a formação de colunas.
from pandas import Series
```

```python
# Criando uma variável com valores a serem expostos em colunas.
myName = Series(['Álvaro', 'Lucas', 'Fernandes'])
myName # Printando.
```

```python
 #Qual o tipo da variável?
type(myName)
```

```python
# Adicionando os indices.
myName = Series(['Álvaro', 'Lucas', 'Fernandes'], index = ['Primeiro', 'Segundo', 'Terceiro'])
myName # Printando.
```

```python
# Adicionando mais valores e seus respectivos indices.
newmyName = myName.append(Series(['De', 'Ananias'], index = ['Quarto', 'Quinto']))
print(newmyName) # Printando.
```

```python
# Retornando o valor pelo indice.
newmyName['Primeiro']
```

```python
#Criando uma série de dados passando um dicionário como parâmetro
#Empresas mais valiosas do mundo (em bilhoes de dólares)

dict = {'Apple': 408.25,'Amazon': 249.25, 'Microsoft': 210.19,
        'Google': 196.81, 'Samsung':74.36}
```

```python
# Criando uma série temporal a partir de um dicionário
obj = Series(dict)
print(obj)
```

```python
# Criando uma lista com empresas valiosas
empresas = ['Apple', 'Amazon', 'Microsoft', 'Google', 'Coca-cola']
```

```python
# Criando uma serie e usando uma lista como indice
obj1 = Series(dict, index=empresas)
print(obj1)
```

```python
# Verificando se algum indice tem valor zerado
obj1.isnull()
```
